```python
from diffusers import DiffusionPipeline
from PIL import Image
import requests
from io import BytesIO

# Load the model pipeline
pipeline = DiffusionPipeline.from_pretrained("lambdalabs/sd-image-variations-diffusers")

# Load an image from a URL or your local system
url = "https://path_to_your_image.jpg"  # Replace with your image URL or local path
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Generate image variations
num_samples = 4
results = pipeline([image] * num_samples, guidance_scale=3)

# Save the output images
output_dir = "outputs/"
for idx, img in enumerate(results["images"]):
    img.save(f"{output_dir}result_{idx}.jpg")
```
